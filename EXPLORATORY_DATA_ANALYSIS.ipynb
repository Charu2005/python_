**Data of marks of hindi **

```
import pandas as pd
import matplotlib.pyplot as plt


# Step 1: Upload the CSV file
path= "/content/drive/MyDrive/Colab Notebooks/hindi.csv"

# Step 2: Load the CSV into a DataFrame
df = pd.read_csv(path)

# Step 3: Plot using df.plot() with subplots
df.plot( kind='line')

# Step 4: Show the plot
plt.tight_layout()  # This will ensure that subplots dont overlap
plt.show()
```
